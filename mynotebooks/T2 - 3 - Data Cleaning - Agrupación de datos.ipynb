# Agregación de datos por categoría

```python
import numpy as np
import pandas as pd 
```

```python
gender = ["Male","Female"]
income = ["Low Class","Middle Class","High Class"]
```

```python
n = 500

gender_data = []
income_data = []

for i in range(0,500):
    gender_data.append(np.random.choice(gender))
    income_data.append(np.random.choice(income))
```

```python
gender_data[1:10]
```

```python
income_data
```

```python
# Z = N(0,1)
# N(m, s) -> m + s * Z
# Normal(media, desviación típica) -> media + desviación típica * (Normal de media 0 y desviación típica 1)
height = 160 + 30 * np.random.randn(n)
weight = 75 + 25 * np.random.randn(n)
age = 30 + 12 * np.random.randn(n)
age_trunc = np.trunc(age)
income = 18000 + 3000 * np.random.randn(n)
```

```python
data = pd.DataFrame(
    {
        "Gender" : gender_data,
        "Economic Status" : income_data,
        "Height" : height,
        "Weight" : weight,
        "Age" : age_trunc,
        "Income" : income
    }
)
```

```python
data.head()
```

## Agrupación de datos

```python
grouped_gender = data.groupby("Gender")
```

```python
grouped_gender.groups
```

```python
for names, groups in grouped_gender:
    print(names)
    print(groups)
```

```python
grouped_gender.get_group("Female")
```

```python
# Agrupación por más de una categoría
# Ejemplo: Gender & Economic Status
double_group = data.groupby(["Gender","Economic Status"])
```

```python
len(double_group)
```

```python
for names, groups in double_group:
    print(names)
    print(groups)
```
