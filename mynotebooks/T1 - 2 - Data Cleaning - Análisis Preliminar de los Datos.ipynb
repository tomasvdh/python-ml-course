## Missing values

```python
# Comprobar si todos los campos de un valor, son nulos
# True = Null, False = NotNull
pd.isnull(data["body"])
```

```python
# Comprobar si todos los campos de un valor, son no nulos
# True = NotNull, False = Null
pd.notnull(data["body"])
```

```python
# Sumamos todos los True y los False para saber cuantos datos nulos hay
# True = 1, False = 0
print("Hay",pd.isnull(data["body"]).values.ravel().sum(),"datos nulos.")
```

```python
# Sumamos todos los True y los False para saber cuantos datos no nulos hay
# True = 1, False = 0
print("Hay",pd.notnull(data["body"]).values.ravel().sum(),"datos no nulos.")
```

### Los valores que faltan en un data set pueden venir por dos razones:
* Extracción de los datos: Cuando se extraen los datos de una base datos, por ejemplo, se pueden incorporar valores faltantes a nuestro dataset debido a incompatibilidades entre el servidor de la base de datos y el proceso de extracción, o devido a la generación del csv o el fichero no es capaz de gestionar ese tipo de dato de forma correcta. En estos casos el valor no es que falte, sino que no puede ser incorporado al fichero. Se puede solucionar optimizando el proceso de extracción y haciendo un scraping, o una obtención de los datos más precisa. 
* Recolección de los datos: En algunos casos los datos no se pueden introducir en la base de datos por no haberlos podido recibir de la forma en la cual han sido recolectados.

## Borrado de valores que faltan

```python
# Borrar todas las entradas de una fila donde están los NA
# axis = 0 para borrar las filas, axis = 1 para borrar las columnas
# how="all" borra toda la fila, si todas las columnas contienen valores NA
data.dropna(axis=0, how="all")
```

```python
# Forward fill. Reemplaza el valor NA con el valor más cercano hacia adelante
data5["age"].fillna(method="ffill")
```

```python
# Backward fill. Reemplaza el valor NA con el valor más cercano hacia atrás
data5["age"].fillna(method="backfill")
```
